```python
#importing the liibraries required for this project . Panda's Library to read the data ,Numpy Library for calculations.
import numpy as np  
import pandas as pd  
from sklearn.metrics.pairwise import cosine_similarity  
from sklearn.feature_extraction.text import CountVectorizer  
```

```python
#loading the dataset.
df = pd.read_csv(r'C:\Users\chhav\Downloads\recipes.csv')
df.head()
```

```python
#selecting the columns which are relevant for our model
df=df[["recipe_name",'ingredients','directions']]  
```

```python
df
```

```python
df.duplicated().sum()
```

```python
#removing the duplicates.
df=df.drop_duplicates('recipe_name') 
```

```python
df = df.reset_index(drop=True)
```

```python
#deleting the rows containing the null values.
df = df.dropna()  
```

```python
df
```

```python
# Preprocess the data
df['recipe_info'] = df['ingredients'] + ' ' + df['directions']
```

```python
count_vectorizer = CountVectorizer(stop_words='english')
count_matrix = count_vectorizer.fit_transform(df['recipe_info'])
```

```python
def get_recommendations(user_input, n=10):
    # Transform user input into Count vector
    user_count = count_vectorizer.transform([user_input])

    # Calculate cosine similarity between user input and recipes
    cosine_similarities = cosine_similarity(user_count, count_matrix).flatten()

    # Get indices of top n similar recipes
    top_indices = cosine_similarities.argsort()[:-n-1:-1]
    
     # Set options to display full content in each column
    pd.set_option('display.max_colwidth', None)

    # Return the names and directions of the top recommended recipes
    recommendations = df.loc[top_indices, ['recipe_name', 'ingredients', 'directions']]
    return recommendations
```

```python
user_input = input("Enter the ingredients: ")
recommendations = get_recommendations(user_input)
for index, row in recommendations.iterrows():
    print(f"Recipe Name: {row['recipe_name']}\n")
    print(f"Ingredients: {row['ingredients']}\n")
    print(f"Directions: {row['directions']}")
    print("\n" + "-"*50 + "\n")
```

```python
import pickle
pickle.dump(get_recommendations, open('recipe.pkl','wb'))
```
